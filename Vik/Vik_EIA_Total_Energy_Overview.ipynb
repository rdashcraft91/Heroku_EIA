```python
import requests
import json
import pymongo
from APIKEY import eia_api
```

```python
### Energy Production ###
energy_sources = ['Coal','Crude Oil','Geothermal','Hydroelectric','Natural Gas (Dry)','Natural Gas (Liquid)','Nuclear','Solar','Total Biomass','Wind']
energy_source_keys = ['TOTAL.CLPRBUS.A','TOTAL.PAPRBUS.A','TOTAL.GETCBUS.A','TOTAL.HVTCBUS.A','TOTAL.NGPRBUS.A','TOTAL.NLPRBUS.A','TOTAL.NUETBUS.A','TOTAL.SOTCBUS.A','TOTAL.BMPRBUS.A','TOTAL.WYTCBUS.A']

url = "http://api.eia.gov/series/?api_key="
series = '&series_id='
query_urls = []
energy_production = {"_id":'production'}
index = 0

for source in energy_source_keys:
    query_url = url + eia_api + series + source
    response = requests.get(query_url)
    response_json = response.json()
    energy_production[energy_sources[index]] = response_json
    index += 1
    
    
print(energy_production)
```

```python
### Energy Consumption ###
energy_sources = ['Coal','Geothermal','Hydroelectric','Natural Gas','Nuclear','Petroleum','Solar','Total Biomass','Wind']
energy_source_keys = ['TOTAL.CLTCBUS.A','TOTAL.GETCBUS.A','TOTAL.HVTCBUS.A','TOTAL.NNTCBUS.A','TOTAL.NUETBUS.A','TOTAL.PMTCBUS.A','TOTAL.SOTCBUS.A','TOTAL.BMTCBUS.A','TOTAL.WYTCBUS.A']

url = "http://api.eia.gov/series/?api_key="
series = '&series_id='
query_urls = []
energy_consumption = {"_id":'consumption'}
index = 0

for source in energy_source_keys:
    query_url = url + eia_api + series + source
    response = requests.get(query_url)
    response_json = response.json()
    energy_consumption[energy_sources[index]] = response_json
    index += 1
    
print(energy_consumption)
```

```python
### Upload data to MongoDB ###
# Connect to database
connection_string = 'mongodb://localhost:27017'
client = pymongo.MongoClient(connection_string)
# Create database
db = client.eia_db
```

```python
total_energy_collection = db.total_energy
total_energy_collection.insert_one(energy_production)
total_energy_collection.insert_one(energy_consumption)
```
